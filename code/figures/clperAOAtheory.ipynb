```python
import pandas
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import seaborn as sb
sb.set_style('whitegrid')
sb.set_palette('deep')

colnames = ['alpha1','CL1','v3','alpha2','CL2','v6','alpha3','CL3','v9','v10']
data = pandas.read_csv('Polar_Graph_1.csv',names=colnames,header=None)
#data["CL2"] = data.CL2.astype(float).fillna(0.0)
print(data.dtypes)
data = data.convert_objects(convert_numeric=True)
print(data.dtypes)

plt.figure(figsize=(10, 7.5))  
ax = plt.subplot(111)  
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.plot(data['alpha2'],data['CL2'], '-o',label='MSHD',lw=2)
plt.plot(data['alpha3'],data['CL3'], '-o',label='Selig 1223',lw=2)
plt.plot(data['alpha1'],data['CL1'], '-o',label='Eppler 423',lw=2)
plt.xticks(fontsize=14) 
plt.xlabel('Angle of Attack [$^\circ$]', fontsize=14)
plt.ylabel('Lift coefficient [$C_L$]', fontsize=14)
plt.legend(loc=3,fontsize=14, bbox_to_anchor=(0., 1.02, 1., .102), ncol=3, mode='expand')

#cl20 = plt.plot(AOAsweep, CL20, label = '$u_\infty$ = $20$ ms$^{-1}$', lw=2) 
#cl40 = plt.plot(AOAsweep, CL40, label = '$u_\infty$ = $40$ ms$^{-1}$', lw=2) 


plt.savefig('../../report/figures/clperAOAtheory.png', bbox_inches='tight')
```
