```python
import matplotlib.pyplot  as plt
import matplotlib.patches as mpatches
import numpy as np
import seaborn as sb
sb.set_style('whitegrid')
sb.set_palette('deep')

mu = 1.5
g  = 9.82
m  = 300
CL0 = 0
CL4 = 1
CL1 = 2.05
CL2 = 2.4
CL3 = 2.6
A  = 0.99
rho = 1.225

r = np.linspace(3, 50, 1000)  # 100 evenly-spaced values from 0 to 50
v0 = np.sqrt(2*mu*m*g*r/(m-mu*CL0*rho*r*A))
v1 = np.sqrt(2*mu*m*g*r/(m-mu*CL1*rho*r*A))
v2 = np.sqrt(2*mu*m*g*r/(m-mu*CL2*rho*r*A))
v3 = np.sqrt(2*mu*m*g*r/(m-mu*CL3*rho*r*A))
v4 = np.sqrt(2*mu*m*g*r/(m-mu*CL4*rho*r*A)) 

plt.figure(figsize=(10, 7.5))  
ax = plt.subplot(111)    
  

p3 = plt.plot(r, v3-v0, label = '$C_L$ = 2.6', lw=2) 
p2 = plt.plot(r, v2-v0, label = '$C_L$ = 2.4', lw=6)
p1 = plt.plot(r, v1-v0, label = '$C_L$ = 2.05', lw=2)
p4 = plt.plot(r, v4-v0, label = '$C_L$ = 1.0', lw=2)   
p0 = plt.plot(r, v0-v0, label = '$C_L$ = 0', lw=2)
plt.legend(loc=3,fontsize=14, bbox_to_anchor=(0., 1.02, 1., .102), ncol=5, mode='expand')

plt.axis([0, 50, -1, 45])
plt.yticks(fontsize=14)    
plt.xticks(fontsize=14) 
plt.xlabel('Turn radius $[m]$', fontsize=14)
plt.ylabel('Relative increas $[\%]$', fontsize=14)

plt.savefig('../../report/figures/turnspeedperCLrelative.png', bbox_inches='tight')
```
